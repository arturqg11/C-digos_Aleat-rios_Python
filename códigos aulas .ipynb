```python
def logistica(x1, r, i):
    x = x1
    pop = []
    pop.append(x)
    
    for t in range(i):
        x2 = r * x * (1 - x)
        
        x = x2
        pop.append(x2)
    return pop
```

```python
import random

def lista_ran(t):
    lista = []
    
    for i in range(t):
        x = random.randint(-1000, 1000)
        
        lista.append(x)
        
    return lista

lista = lista_ran(100)

def bubble(lista):
    lista_sort = lista
    
    for i in range(len(lista) - 1):
        for c in range(len(lista) - 1):
            if lista_sort[c] > lista_sort[c+1]:
                lista_sort[c], lista_sort[c+1] = lista_sort[c+1], lista_sort[c]
                
    return lista_sort

print(bubble(lista))
```

```python
import numpy as np

def isOrtogonal(x, y):
    u = np.array(x)
    v = np.array(y)
    p = 0
    
    for i in range(len(u)):
        p += u[i] * v[i]
        
    if p == 0:
        return True
        
    else:
        return False
    
isOrtogonal([1,0,0], [0,1,0])
```

```python
import numpy as np

def centroDeMassa(p):
    "CentroDeMassa recebe uma lista de elementos do tipo (coord_x, cooor_y, massa)"
    
    XCM = 0
    YCM = 0
    M = 0
    pontos = np.array(p)
    
    for i in range(len(pontos)):
        M += pontos[i][2]
        
    for x in range(len(pontos)):
        XCM += pontos[x][0] * pontos[i][3]
        YCM += pontos[x][1] * pontos[i][3]
        
    XCM, YCM = XCM/M, YCM/M 
    
    return np.array(XCM, YCM)
```

```python
import matplotlib.pyplot as  plt
import numpy as np

def Circulo(raio):
    N = 300
    arr = np.linspace(-np.pi, np.pi, N)
    
    for i in arr:
        x = raio * np.cos(i)
        y = raio * np.sin(i)
        
        plt.plot(x, y, "g.-")
        
def Triangulo(lado):
    r = (np.sqrt(3) / 3) * lado
    
    p1 = [0, r]
    p2 = [(lado/2), -r*np.sin(np.pi/6)]
    p3 = [-(lado/2), -r*np.sin(np.pi/6)]
    Px = [p1[0], p2[0], p3[0], p1[0]]
    Py = [p1[1], p2[1], p3[1], p1[1]]   
    
    plt.plot(Px, Py, ".-")
    return r

def Desenho(inicial, itera):
    valor = inicial
    
    for i in range(itera):
        Circulo(Triangulo(valor))
        
        valor *= 2  
        
Desenho(5, 5)
```
